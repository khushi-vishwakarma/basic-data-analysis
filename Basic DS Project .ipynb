# Q.15 What is the distribution of Netflix TV shows/movies over time by release date? OR  Are shows or movies which is more popular show by visulaization? 

```python
data.head(2)
```

```python
# merge rows by release date
q3_data = data.groupby(['Year'],as_index=False)['Show_Id'].count() 
q3_data = q3_data.rename(columns={'Show_Id':"count"})

# visualize the data as a line plot
x = q3_data['Year']
y = q3_data['count']
plt.figure(figsize=(9, 3))
plt.plot(x, y)
plt.xlabel("Release Year")
plt.ylabel("Number of Movies/Shows")

plt.show()
```

```python
data.head(2)
```

```python
q2_data = data.groupby(['Category'],as_index=False)['Show_Id'].count() 
q2_data = q2_data.rename(columns={'Show_Id':"count"})

# visualize the data as a bar plot
t = q2_data['Category']
count = q2_data['count']
plt.figure(figsize=(9, 3))
plt.bar(t, count)
plt.xlabel("Category")
plt.ylabel("Count")
```
